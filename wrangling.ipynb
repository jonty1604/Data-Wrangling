# Concat and append

```python
df1= {
    's_id' : ['1','2','3'],
    'month' : ['jan','feb','mar'],
    'sales' : ['1200','1300','1400']
}
```

```python
df1=pd.DataFrame(df1,columns = ['s_id','month','sales'])
df1
```

```python
df2= {
    's_id' : ['4','5'],
    'month' : ['apr','may'],
    'sales' : ['4000','3500']
}
```

```python
df2=pd.DataFrame(df2,columns = ['s_id','month','sales'])
df2
```

```python
df3={
    'sales_threshold': ['yes','no','yes','yes','no'],
    'bonus_threshold': ['no','no','no','yes','yes']
}
df3=pd.DataFrame(df3, columns = ['sales_threshold','bonus_threshold'])
df3
```

```python
df_row= pd.concat([df1,df2],ignore_index=True)
df_row
```

```python
df_full= pd.concat([df_row,df3],axis=1)
df_full
```

```python
df_append= df1.append(df2).append(df3)
df_append
```

# Merges

```python
daf1=pd.DataFrame({
    'product': ['a','b','c'],
    'price': [100,300,150]
})
daf2=pd.DataFrame({
    'div': ['x','y','z'],
    'prod': ['b','c','a']
})
daf_merge=pd.merge(daf1,daf2,left_on='product',right_on='prod')
daf_merge
```

```python
daf_merge=daf_merge.drop('prod',axis=1)
daf_merge
```

```python
daf_indexmerge=pd.merge(daf1,daf2,left_index=True,right_index=True)
daf_indexmerge
```

```python
daf_concatcolumn=pd.concat([daf1,daf2],axis=1)
daf_concatcolumn
```

# Suffixes


```python
dfs1=pd.DataFrame({
    'prod':['a','b','c'],
    'price':[100,300,250]
})
dfs2=pd.DataFrame({
    'prod':['a','b','c'],
    'price':[600,200,300]
})
dfs_merge=pd.merge(dfs1,dfs2,on='prod')
dfs_merge
```

```python
dfs_merge=pd.merge(dfs1,dfs2,on='prod',suffixes=['_lp','_rp'])
dfs_merge
```

# Update

```python
dup1=pd.DataFrame({
    'name':['a','b'],
    'val':0
})
dup2=pd.DataFrame({
    'name':['b','a'],
    'val':[10,20]
})
dup1.update(dup2)
dup1
```

```python
dup2.combine_first(dup1)
dup2
```

# Data Normalization

```python
#Simple feature scaling . f/fmax
kdf['Age']=kdf['Age'].astype(int)
```

```python
kdf['Age'].max()
```

```python
kdf_dup=kdf.copy()
kdf_dup['Age']=kdf['Age']/kdf['Age'].max()
kdf_dup.head()
```

```python
#Min-Max : f-fmin / fmax-fmin
kdf_dup['DistanceFromHome']=kdf_dup['DistanceFromHome'].astype(int)
```
